```python
import pandas as pd
import numpy as np
from sklearn import svm
import matplotlib as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, KFold
%matplotlib inline
```

```python
# List to denote custom field names
fields = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
data = pd.read_csv('housing.csv', names = fields, delim_whitespace = True)
data.head()
```

```python
# Divide data by X = Features, y = Labels
X = data.drop(['MEDV','CHAS'],1)
y = data['MEDV']

print('Features shape: ',X.shape)
print('Classes shape: ',y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .25, random_state = 3)

print('Train Data shape: ',X_train.shape)
print('Train Classes shape: ',y_train.shape)
print("")
print('Test Data shape: ',X_test.shape)
print('Test Classes shape: ',y_test.shape)
# scaler= MinMaxScaler(feature_range = (0,1))
scaler = StandardScaler()
# Scaled Features to feed to the classifier
X_scaled = scaler.fit_transform(X_train)
Xt_scaled = scaler.transform(X_test)
```

```python
# sns.pairplot(data)
```

```python
clf = svm.SVR(kernel = 'linear')
clf
```

```python
clf.fit(X_scaled,y_train)
print('Cross Val. Score: ', cross_val_score(clf, Xt_scaled, y_test).mean())
print('Model Accuracy Score: ', clf.score(Xt_scaled, y_test))
```

```python
KF = KFold(n_splits = 3)
```

```python
pred = clf.predict(Xt_scaled[1:5])
print('Prediction: ', pred)
print('\n')
for i in y_test[1:5]:
    print('Original Price: ',i)
print('')
# print('Coeffitients: \n', clf.coef_)
```
